```python
# Попробуем найти хеши тех, кто проголосовал, среди хешей тех, кто есть в Google Контактах

def findByHash(hash):
    try:
        return allMIPThashes.index(hash)
    except:
        return "Unknown"

whoVoted = [findByHash(hash) for hash in hashesWhoVoted]
print(whoVoted)
```

```python
# Число нераспознанных
sum([i == "Unknown" for i in whoVoted])
```

На предыдущем шаге мы получили массив `whoVoted` – это номера строчек в `contacts.csv`.

Теперь мы делаем таблицу, в которой вместо номера будет ФИО и электронная почта.

```python
def findByIndex(columnName, index):
    """Ищет ячейку в contacts"""
    
    if index == "Unknown":
        return "Unknown"
    else:
        return contacts[columnName][index]
```

```python
decryptedTable = pandas.DataFrame({
        "Name" : [findByIndex("Name", i) for i in whoVoted],
        "Email": [findByIndex("E-mail 1 - Value", i) for i in whoVoted],
        "Hash" : hashesWhoVoted,
        "Vote" : votes,
    })
```
